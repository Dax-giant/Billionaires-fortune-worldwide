```python
# Importing all liberies we need
# 1.Importing the requests lib for requesting throw http protocol
import requests 
# 2.Importing Pandas lib for prossing the data
import pandas as pd
import numpy as np
```

```python
# Importing libraries for data visualisation
import matplotlib.pyplot as plt
# This line is for plotting my charts everytime i need them, without calling plt.show() method in matplotlib 
%matplotlib inline 
```

```python
# This URL is the endpoint of forbes.com API
url = 'https://www.forbes.com/forbesapi/person/rtb/0/-estWorthPrev/true.json?fields=rank,uri,personName,lastName,gender,source,industries,countryOfCitizenship,birthDate,finalWorth,estWorthPrev,imageExists,squareImage,listUri'
```

```python
# We pulled the data from Forbes.com API using request
# I used the .json() method to return json object wiche is a dict
pulled_data = requests.get(url).json()
print(pulled_data)
```

```python
# We get our pulleded data as python dict
type(pulled_data)
```

```python
# Our pulled data is now a nested dict, so we need to access keys to get the data we need
# Let's access first key of the dict whiche is 
dictt = pulled_data['personList']
# Let's acces second key to get 
data_list = dictt['personsLists']
```

```python
# Let's use Pandas to build our dataframe object
df = pd.DataFrame(data_list)
df.head(10)
```

```python
# Getting some infos about our data
df.info()
```

```python
# As u can see in birthDate column we have a problem with type of data 
df['birthDate']
```

```python
# It's time to make our column change from float64 to datetime object to make it more sense
# Notice that set unit to milisecond to get date from 
df['birthDate'] = pd.to_datetime(df['birthDate'], unit='ms')
df['birthDate']
```

```python
# Let's see what our dataframe looks like
df.head()
```

```python
# In this step i will calculate the age of every person we have 
# For that i'm importing datetime module so i can get today's date
from datetime import date, datetime
today = date.today()
today = np.datetime64(today)
print("Today is:", today)
```

```python
# Creating function age calculator
def Age_cal(birthDate):
    birthDate = np.datetime64(birthDate)
    age = today - birthDate
    return age
```

```python
# I create new column called Stagestep just as middle step to save my processed data in 
# I found there's a NAN values, so i used fillna method to replace them with today's date
df['Stagestep']=df['birthDate'].fillna(today)
```

```python
# I apply the age calculator function to column
df['Stagestep'] = df['Stagestep'].apply(Age_cal)
# As u can see, i get the resualts as days unit
df['Stagestep']
```

```python
# I apply the function from pandas timedelta64 and astype(int) methode to get days as integer number so ican perfrom calculation in the next step
df['Stagestep']=df['Stagestep'].apply(lambda x:np.timedelta64(x, 'D').astype(int))
```

```python
# I create this function to return years from days by dividing on 356.25 days 
def get_years(days):
    return int(days / 365.25)
```

```python
# I apply get_years function to 'Stagestep' to return years as intger numbers
df['Stagestep'] = df['Stagestep'].apply(get_years)
df['Stagestep']
```

```python
# I create function to return ages as string + "Years old"
def age_str(age):
    return str(age) + " Years old"
df["Age"] = df['Stagestep'].apply(age_str)
df["Age"]
```

```python
# Creating new dataframe with specfiques columns from our df + new column named Age
df_data = df[['rank', 'personName', 'finalWorth', 'Age', 'source', 'countryOfCitizenship']].copy()
df_data.head(10)
```

```python
# Changing the names of columns
col_dict = {'rank':'Rank', 'personName':'Full Name', 'birthDate':'Birth Date', 'finalWorth':'Net Worth', 'source':'Source', 'countryOfCitizenship':'Country/Territory'}
# Creating a dict with new columns names to pass them to rename method
df_data.rename(columns=col_dict, inplace=True)
```

```python
# Creating new column to calculate the percentage of each billionaire from global fourtune
df_data['Per from globalfourtune (%)'] = ((df_data['Net Worth'] / df_data['Net Worth'].sum())*100).round(decimals=2)
df_data
```

```python
# I creat a top 10 billionaires + others billionaires so, i can perfome plotting 
# In this step i creat first dataframe to get top ten
df_billionaires = df_data[:10][['Full Name', 'Net Worth', 'Per from globalfourtune (%)']]
# In this one i creat other billionaires dataframe 
other_billionaires = df_data[10:][['Full Name', 'Net Worth', 'Per from globalfourtune (%)']]
other_billionaires = pd.DataFrame(data={'Full Name' : 'Other Billionaires',
                                        'Net Worth' : df_data[10:]['Net Worth'].sum(),
                                        'Per from globalfourtune (%)' : df_data[10:]['Per from globalfourtune (%)'].sum()}, 
                                  index=[0],
                                 )
# Here i concate the tow dataframes
top_billionaires = pd.concat([df_billionaires, other_billionaires]).reset_index(drop=True)
top_billionaires
```

```python
top_billionaires['Net Worth']
```

```python
# I want to know how many country i have in my dataset so, i can get a preview of my worldwild plot 
# I used the unique() methode to count just the unique countrys and i used the method size so i can get the value 71 countries
print(str(df_data['Country/Territory'].unique().size) + " Countries")
```

```python
# We found that we have we have more than 13751782.162 Billions distributed through 71 countries
# So let's group by countries to found how much fortune we have for each country
grouped_fortune = df_data.groupby(['Country/Territory'], as_index = False).sum().sort_values('Net Worth', ascending=False)
# TOP 5 countries are US, CHINA, INDIA, Germany and RUSSIA
grouped_fortune = grouped_fortune[['Country/Territory', 'Net Worth', 'Per from globalfourtune (%)']].reset_index()
grouped_fortune.drop('index', axis=1, inplace=True)
grouped_fortune = grouped_fortune[['Country/Territory', 'Net Worth', 'Per from globalfourtune (%)']]
grouped_fortune
```

```python
# I did this step to count how many billionaires are in the our data and how many billions are there
billionaires = df_data['Full Name'].size
billiones = df_data['Net Worth'].sum()
print("We have {} billionaires with worth of {} Billiones".format(billionaires, round(billiones)))
```

```python
##### Let's Start our data visualsation by plotting the grouped fortune for each country
# In this step i will create to new small dataframes the first is top 10 countries and the second is for the rest of countries
# the first dataframe
df_top = grouped_fortune[0:10] 
# Creating a place holders for two values 
net_worth = grouped_fortune[10:]['Net Worth'].sum().round()
pfg_data = grouped_fortune[10:]['Per from globalfourtune (%)'].sum().round()
# the second dataframe
df_others = pd.DataFrame(data={'Country/Territory': 'Other countries', 'Net Worth': net_worth, 'Per from globalfourtune (%)' : pfg_data}, index=[0])
# Concatenate the two dataframes
grouped_fortune = pd.concat([df_top, df_others]).reset_index(drop=True)
grouped_fortune = grouped_fortune[['Country/Territory', 'Net Worth', 'Per from globalfourtune (%)']] 
grouped_fortune
```

```python
# In next step i ll rename the columns with using B and % as units to cleare that my numbers are in billions and percentage
grouped_fortune.rename(columns={'Net Worth':'Net Worth (Bi)'}, inplace=True)
top_countries = grouped_fortune
top_countries
```

```python
# Pie chart for top 10 countries 
# I build a list of zeros with the length of countries,so i can pass a value of 0.1 to explode last element "other countries"
explode = [0] * len(top_countries['Net Worth (Bi)'])
explode[-1] = 0.1
fig1, axe1 = plt.subplots(figsize=(10,10))
axe1.pie(x=top_countries['Net Worth (Bi)'], 
         labels=top_countries['Country/Territory'] ,
         explode = explode,
         autopct='%1.2f%%',
         pctdistance=0.9,
         wedgeprops={"edgecolor":"white", 'linewidth': 2, 'antialiased': True})
axe1.set_title("Fortune distribution across countries")
fig1.tight_layout()
```

```python
# Let's Plot top 10 Billionaires using horizontal Bar chart builtin in matplotlib library to see how everyone of them shares
# Specifying the data we need
name = top_billionaires["Full Name"]
shares = top_billionaires["Net Worth"]

# Figure Size
fig2, axe2 = plt.subplots(figsize=(20, 7))
# Creating horizontal Bar Plot
axe2.barh(name, shares)
# Inverting the y axe to Showing top values
axe2.invert_yaxis() 
# Add annotation to bars
for i in axe2.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
# Setting tiltles
axe2.set_xlabel("Shares in billions")
axe2.set_ylabel("Full Name")
axe2.set_title("Shares of Billionaires from global fortune")
fig2.tight_layout()
```

```python
#Question: What's the industry makes more money? old saying that the petrol companies are the top, let's find out?
# So let's Group by industries to know which industry makes more money
df['industries'].head(20)
# I found that every element of 'industries' column is an list object
# So i need to make them str to performe a group by using pandas
```

```python
# I will create a function to handle that 
# I found that the column contains some values not a list, so i control that using if statement
def list_to_str(lst):
    if type(lst) == list:
        mylist = ",".join(lst)
        return str(mylist)
    else:
        return "Unknown"
```

```python
df['Industries'] = df['industries'].apply(list_to_str)
df_data['Industries'] = df['Industries']
```

```python
groupby_Industries = df.groupby(['Industries'], as_index = False).sum().sort_values('finalWorth', ascending=False)
groupby_Industries[['Industries','finalWorth']].reset_index(drop=True)
# So we found out the answer for our question, is invest as much you can in Technology industry
```
